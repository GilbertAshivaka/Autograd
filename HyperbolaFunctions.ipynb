```python
import numpy as np
import matplotlib.pyplot as plt

# Define x range
x = np.linspace(-3, 3, 400)

# Create masked arrays to handle singularities
x_coth_csch = x[abs(x) > 0.05]  # Avoid points too close to zero

# Compute hyperbolic functions
sinh_x = np.sinh(x)
cosh_x = np.cosh(x)
tanh_x = np.tanh(x)
coth_x = 1 / np.tanh(x_coth_csch)
sech_x = 1 / np.cosh(x)
csch_x = 1 / np.sinh(x_coth_csch)

# Create the plots
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

# Function to set common properties for all subplots
def setup_subplot(ax, ylim=(-3, 3)):
    ax.grid(True, linestyle='--', alpha=0.3)
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_xlim(-3, 3)
    ax.set_ylim(ylim)

# Plot each function
ax[0, 0].plot(x, sinh_x, label=r'$\sinh(x)$', color='b') # Corrected label
setup_subplot(ax[0, 0], (-5, 5))
ax[0, 0].legend()
ax[0, 0].set_title('Hyperbolic Sine')

ax[0, 1].plot(x, cosh_x, label=r'$\cosh(x)$', color='r') # Corrected label
setup_subplot(ax[0, 1], (0, 5))
ax[0, 1].legend()
ax[0, 1].set_title('Hyperbolic Cosine')

ax[0, 2].plot(x, tanh_x, label=r'$\tanh(x)$', color='g') # Corrected label
setup_subplot(ax[0, 2], (-1.5, 1.5))
ax[0, 2].legend()
ax[0, 2].set_title('Hyperbolic Tangent')

ax[1, 0].plot(x_coth_csch, coth_x, label=r'$\coth(x)$', color='purple') # Corrected label
setup_subplot(ax[1, 0], (-5, 5))
ax[1, 0].legend()
ax[1, 0].set_title('Hyperbolic Cotangent')

ax[1, 1].plot(x, sech_x, label=r'$\mathrm{sech}(x)$', color='orange') # Corrected label using \mathrm
setup_subplot(ax[1, 1], (0, 1.2))
ax[1, 1].legend()
ax[1, 1].set_title('Hyperbolic Secant')

ax[1, 2].plot(x_coth_csch, csch_x, label=r'$\mathrm{csch}(x)$', color='brown') # Corrected label using \mathrm
setup_subplot(ax[1, 2], (-5, 5))
ax[1, 2].legend()
ax[1, 2].set_title('Hyperbolic Cosecant')

# Add overall title
fig.suptitle('Hyperbolic Functions', fontsize=16, y=1.02)

# Adjust layout
plt.tight_layout()
plt.show()
```
