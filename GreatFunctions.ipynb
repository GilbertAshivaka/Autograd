```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

def swish(x):
    return x * sigmoid(x)

def softplus(x):
    return np.log(1 + np.exp(x))

def gaussian(x, mu=0, sigma=1):
    return np.exp(-((x - mu) ** 2) / (2 * sigma**2))

# Create an array of x values
x = np.linspace(-10, 10, 400)

# Set up the plot grid
plt.figure(figsize=(12, 10))

# Plot each function in a subplot
plt.subplot(3, 3, 1)
plt.plot(x, sigmoid(x), color='blue')
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(3, 3, 2)
plt.plot(x, tanh(x), color='green')
plt.title("Tanh")
plt.grid(True)

plt.subplot(3, 3, 3)
plt.plot(x, relu(x), color='red')
plt.title("ReLU")
plt.grid(True)

plt.subplot(3, 3, 4)
plt.plot(x, leaky_relu(x), color='purple')
plt.title("Leaky ReLU")
plt.grid(True)

plt.subplot(3, 3, 5)
plt.plot(x, swish(x), color='orange')
plt.title("Swish")
plt.grid(True)

plt.subplot(3, 3, 6)
plt.plot(x, softplus(x), color='brown')
plt.title("Softplus")
plt.grid(True)

plt.subplot(3, 3, 7)
plt.plot(x, gaussian(x), color='magenta')
plt.title("Gaussian")
plt.grid(True)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
