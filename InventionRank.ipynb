```python
import matplotlib.pyplot as plt
import numpy as np

# Define the categories (criteria)
categories = [
    "Transformative Impact",
    "Scope of Influence",
    "Longevity",
    "Catalytic Role",
    "Foundational Nature"
]
N = len(categories)

# Compute angles for each axis in the plot (in radians)
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # complete the loop

# Data for the top 12 inventions (scores are on a 1–10 scale)
# The chosen inventions are based on high composite ratings
inventions = {
    "Language": [10, 10, 10, 10, 10],
    "Agriculture": [10, 10, 10, 10, 10],
    "Writing": [10, 10, 10, 10, 10],
    "Electricity": [10, 10, 10, 10, 10],
    "Vaccination": [10, 10, 10, 9, 10],
    "The Wheel": [10, 10, 10, 9, 9],
    "Printing Press": [10, 10, 9, 10, 9],
    "Steam Engine": [10, 10, 9, 10, 9],
    "The Computer": [10, 10, 9, 10, 9],
    "The Internet": [10, 10, 9, 10, 9],
    "Transistor": [10, 10, 9, 10, 9],
    "Battery": [10, 10, 10, 9, 9]
}

# Initialize the radar chart
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories, fontsize=12)

# Set the range for the radial axis
ax.set_rlabel_position(0)
plt.yticks([2,4,6,8,10], ["2","4","6","8","10"], color="grey", size=10)
plt.ylim(0, 10)

# Plot each invention's scores on the radar chart
for invention, scores in inventions.items():
    values = scores + [scores[0]]  # repeat the first score to close the circle
    ax.plot(angles, values, linewidth=2, label=invention)
    ax.fill(angles, values, alpha=0.1)

plt.title("Radar Chart of Top 12 Great Inventions", size=16, y=1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.show()
```
